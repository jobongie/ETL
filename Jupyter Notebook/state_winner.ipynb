```python
import pandas as pd
import os
import requests
import json
import numpy as np
from pprint import pprint
```

```python
state_winner_df = pd.read_csv('data/grouped_data_df.csv')

state_winner_df
```

```python
conditions = [
    (state_winner_df['Trump Adjusted'] > state_winner_df['Clinton Adjusted']),
    (state_winner_df['Trump Adjusted'] < state_winner_df['Clinton Adjusted']),
]

values = ['Trump', 'Clinton']
state_winner_df['Winner'] = np.select(conditions, values)

state_winner_df
```

```python
output_file = os.path.join("data","state_winner_df.csv")
state_winner_df.to_csv(output_file)
```
